# 4 Análise Bivariada

Vamos utilizar de técnicas associativas para entender quais são as variáveis que mais se correlacionam com a nossa variável target (interesse_no_Curso)

interesse_no_Curso x Variaveis Quantitativas

Informatio Value

```python
import numpy as np
import pandas as pd
from typing import Dict, List

def calculate_woe_iv(df: pd.DataFrame, feature: str, target: str) -> pd.DataFrame:
    """
    Calcula Weight of Evidence (WoE) e Information Value (IV) para uma variável.
    
    Args:
        df: DataFrame com os dados
        feature: Nome da variável a ser analisada
        target: Nome da variável target
        
    Returns:
        DataFrame com os cálculos de WoE e IV
    """
    # Agrupa os dados e calcula distribuições
    grouped = df.groupby(feature)[target].agg(['count', 'sum'])
    grouped.columns = ['Total', 'Good']
    grouped['Bad'] = grouped['Total'] - grouped['Good']
    
    # Calcula proporções
    total_good = grouped['Good'].sum()
    total_bad = grouped['Bad'].sum()
    
    # Evita divisão por zero
    if total_good == 0 or total_bad == 0:
        raise ValueError("Total de 'Good' ou 'Bad' é zero. Verifique os dados.")
    
    grouped['Distr_Good'] = grouped['Good'] / total_good
    grouped['Distr_Bad'] = grouped['Bad'] / total_bad
    
    # Calcula WoE e IV, tratando casos onde Distr_Good ou Distr_Bad são zero
    grouped['WoE'] = np.where(
        (grouped['Distr_Good'] > 0) & (grouped['Distr_Bad'] > 0),
        np.log(grouped['Distr_Good'] / grouped['Distr_Bad']),
        0  # Define WoE como 0 se Distr_Good ou Distr_Bad for zero
    )
    
    grouped['IV'] = (grouped['Distr_Good'] - grouped['Distr_Bad']) * grouped['WoE']
    
    # Formata percentuais
    grouped['Distr_Good'] = grouped['Distr_Good'].apply(lambda x: f"{x:.2%}")
    grouped['Distr_Bad'] = grouped['Distr_Bad'].apply(lambda x: f"{x:.2%}")
    grouped['Taxa_Interesse'] = (grouped['Good'] / grouped['Total']).apply(lambda x: f"{x:.2%}")
    
    return grouped

def calculate_all_ivs(df: pd.DataFrame, variables: List[str], target: str) -> Dict:
    """
    Calcula IV para múltiplas variáveis e retorna resultados organizados.
    
    Args:
        df: DataFrame com os dados
        variables: Lista de variáveis para análise
        target: Nome da variável target
        
    Returns:
        Dicionário com resultados e IV total
    """
    results = {}
    iv_scores = {}
    
    for var in variables:
        try:
            result_df = calculate_woe_iv(df, var, target)
            results[var] = result_df
            iv_scores[var] = result_df['IV'].sum()
        except ValueError as e:
            print(f"Erro ao calcular WoE/IV para {var}: {e}")
            results[var] = None
            iv_scores[var] = None
    
    return {'details': results, 'iv_scores': iv_scores}

def prepare_data(df: pd.DataFrame) -> pd.DataFrame:
    """Prepara os dados para análise."""
    df = df.copy()
    
    # Categoriza valor_que_pagaria com as novas faixas solicitadas
    df['valor_que_pagaria_bin'] = pd.cut(
        df['valor_que_pagaria'],
        bins=[0, 50, 90, 110, float('inf')],
        labels=['0-50','50-90','90-110','110+']
    )
    
    return df

def main(df: pd.DataFrame) -> None:
    """Função principal de execução."""
    # Prepara dados
    df_prepared = prepare_data(df)
    
    # Verifica a distribuição de interesse_no_Curso
    print("Distribuição de interesse_no_Curso:")
    print(df_prepared['interesse_no_Curso'].value_counts())
    
    # Define variáveis para análise
    variables = ['sexo', 'faixa_etária', 'escolaridade', 'valor_que_pagaria_bin']
    
    # Calcula WoE e IV
    results = calculate_all_ivs(df_prepared, variables, 'interesse_no_Curso')
    
    # Exibe resultados
    print("=== Information Value (IV) por Variável ===")
    for var, iv_score in results['iv_scores'].items():
        if iv_score is not None:
            print(f"\n{var}: {iv_score:.4f}")
            print("-" * 50)
            print(results['details'][var])
    
    # Classifica poder preditivo
    print("\n=== Classificação do Poder Preditivo ===")
    for var, iv in results['iv_scores'].items():
        if iv is not None:
            power = ""
            if iv < 0.02: power = "Não Preditiva"
            elif iv < 0.1: power = "Fraco"
            elif iv < 0.3: power = "Médio"
            elif iv < 0.5: power = "Forte"
            else: power = "Suspeito"
            print(f"{var}: {iv:.4f} - {power}")

if __name__ == "__main__":
    # Seus dados aqui
    main(df)
```

## Perfil de cliente ideal

```python
import numpy as np
import pandas as pd
from typing import Dict, List

def calculate_woe_iv(df: pd.DataFrame, feature: str, target: str) -> pd.DataFrame:
    """
    Calcula Weight of Evidence (WoE) e Information Value (IV) para uma variável.
    """
    grouped = df.groupby(feature)[target].agg(['count', 'sum'])
    grouped.columns = ['Total', 'Good']
    grouped['Bad'] = grouped['Total'] - grouped['Good']
    
    total_good = grouped['Good'].sum()
    total_bad = grouped['Bad'].sum()
    
    epsilon = 1e-10
    grouped['Distr_Good'] = (grouped['Good'] + epsilon) / (total_good + epsilon)
    grouped['Distr_Bad'] = (grouped['Bad'] + epsilon) / (total_bad + epsilon)
    
    grouped['WoE'] = np.log(grouped['Distr_Good'] / grouped['Distr_Bad'])
    grouped['IV'] = (grouped['Distr_Good'] - grouped['Distr_Bad']) * grouped['WoE']
    
    grouped['Taxa_Interesse'] = (grouped['Good'] / grouped['Total']).apply(lambda x: f"{x:.2%}")
    
    return grouped

def analyze_high_value_customers(df: pd.DataFrame) -> pd.DataFrame:
    """
    Análise específica para clientes dispostos a pagar R$85 ou mais.
    """
    df = df.copy()
    
    # Cria flag para clientes de alto valor (>= R$85)
    df['high_value'] = (df['valor_que_pagaria'] >= 85).astype(int)
    
    # Variáveis para análise
    variables = ['sexo', 'faixa_etária', 'escolaridade']
    
    results = {}
    print("\n=== Análise de Perfil - Clientes R$85+ ===")
    
    for var in variables:
        result = calculate_woe_iv(df, var, 'high_value')
        results[var] = result
        
        print(f"\n{var.upper()}:")
        print("Taxa de conversão para valor >= R$85 por categoria:")
        print(result[['Total', 'Taxa_Interesse']])
        
        # Identifica a categoria mais propensa
        best_category = result.sort_values('WoE', ascending=False).index[0]
        print(f"Categoria mais propensa: {best_category}")
    
    return results

def prepare_data(df: pd.DataFrame) -> pd.DataFrame:
    """Prepara os dados para análise."""
    df = df.copy()
    
    print("\nDistribuição de valor_que_pagaria:")
    print(df['valor_que_pagaria'].describe())
    
    print("\nQuantidade de clientes dispostos a pagar R$85+:")
    print((df['valor_que_pagaria'] >= 85).value_counts())
    
    return df

def main(df: pd.DataFrame) -> None:
    """Função principal de execução."""
    df_prepared = prepare_data(df)
    results = analyze_high_value_customers(df_prepared)
    
    print("\n=== PERFIL IDEAL PARA VALOR >= R$85 ===")
    for var, result_df in results.items():
        # Pega a categoria com maior WoE
        best_category = result_df.sort_values('WoE', ascending=False).index[0]
        conversion = result_df.loc[best_category, 'Taxa_Interesse']
        print(f"{var}: {best_category} (Conversão: {conversion})")

if __name__ == "__main__":
    # Seus dados aqui
    main(df)
```

# Se sim, por qual preço médio? Além disso, é possível entender o perfil dos interessados?




```python
# 1. Filtra o DataFrame para remover as pessoas que indicaram R$50
df_filtered = df[df['valor_que_pagaria'] > 50]

# 2. Exibe a nova distribuição de valor_que_pagaria
print("\nNova distribuição de valor_que_pagaria (após exclusão de R$50):")
print(df_filtered['valor_que_pagaria'].describe())

# 3. Calcula o novo preço médio
novo_preco_medio = df_filtered['valor_que_pagaria'].mean()
print(f"\nNovo preço médio (após exclusão de R$50): R${novo_preco_medio:.2f}")

# 4. Cria flag para clientes de alto valor (>= R$85)
df_filtered['high_value'] = (df_filtered['valor_que_pagaria'] >= 85).astype(int)

# 5. Variáveis para análise
variables = ['sexo', 'faixa_etária', 'escolaridade']

# 6. Realiza a análise de perfil para clientes dispostos a pagar R$85+
results_filtered = {}
print("\n=== Nova Análise de Perfil - Clientes R$85+ (após exclusão de R$50) ===")

for var in variables:
    result = calculate_woe_iv(df_filtered, var, 'high_value')
    results_filtered[var] = result
    
    print(f"\n{var.upper()}:")
    print("Taxa de conversão para valor >= R$85 por categoria:")
    print(result[['Total', 'Taxa_Interesse']])
    
    # Identifica a categoria mais propensa
    best_category = result.sort_values('WoE', ascending=False).index[0]
    print(f"Categoria mais propensa: {best_category}")

# 7. Exibe o perfil ideal para clientes dispostos a pagar R$85+
print("\n=== PERFIL IDEAL PARA VALOR >= R$85 (após exclusão de R$50) ===")
for var, result_df in results_filtered.items():
    # Pega a categoria com maior WoE
    best_category = result_df.sort_values('WoE', ascending=False).index[0]
    conversion = result_df.loc[best_category, 'Taxa_Interesse']
    print(f"{var}: {best_category} (Conversão: {conversion})")
```
